```python
import pandas as pd

df = pd.read_csv("../data/SampleSuperstore.csv")

print("Shape:", df.shape)

print("\nFirst rows:")
print(df.head())

print("\nInformation:")
print(df.info())

print("\nNulls:")
print(df.isnull().sum())

print("\nDuplicates:")
print(df.duplicated().sum())
```

```python
# Delete duplicates
df = df.drop_duplicates()

print("\nShape after cleaning:", df.shape)
print("\nRemaining duplicates:", df.duplicated().sum())
```

```python
# Check for unique values in categorical columns
categorical = ['Ship Mode', 'Segment', 'Country', 'City', 'State', 'Postal Code',
               'Region', 'Category', 'Sub-Category', 'Sales', 'Quantity', 'Discount', 'Profit']

for cat in categorical:
    print(f"\n{cat} ({df[cat].nunique()} unique values):")
    print(df[cat].unique())
```

```python
# Export clean dataset
df.to_csv("../data/processed/Superstore_Cleaned.csv", index=False)

print("Clean dataset exported successfully.")
print("Final shape:", df.shape)
```
